```python
# Filter the rows where "Harvard University" appears
harvard_university_rows = df[df['University Name'] == 'Harvard University']

# Remove specific columns from the filtered rows
df_filtered = harvard_university_rows.drop(['Hostel or Non-hostel', 'Winners of the Year', 'Internship', 'Student Address', 'Student ID'], axis=1)


print(df_filtered)
```

```python
# Filter the rows where "MIT" (case-insensitive) appears
mit_count_rows = df[df['University Name'].str.lower() == 'mit']

# Remove specific columns from the filtered rows
df_filtered = mit_count_rows.drop(['Hostel or Non-hostel', 'Winners of the Year', 'Internship', 'Student Address', 'Student ID'], axis=1)

# Show the resulting DataFrame
print(df_filtered)
```

```python
# Assuming the 'Fees' column contains the fees of the students
max_fees_row = df[df['Fees'] == df['Fees'].max()]

# Showing the row with the maximum fees
print(max_fees_row)
```

```python
# Get the maximum scholarship amount
max_scholarship = df['Scholarship'].max()

# Create a new column 'Max Scholarship' to indicate if the university has the maximum scholarship
df['Max Scholarship'] = df['Scholarship'] == max_scholarship

# Show the resulting DataFrame with the new column
print(df[['University Name', 'Scholarship', 'Max Scholarship']])
```

```python
# Remove leading and trailing spaces from column names
df.columns = df.columns.str.strip()
```

```python
# Set the figure size
plt.figure(figsize=(12, 6))

# Bar plot for Fees
plt.subplot(1, 2, 1)  # (rows, columns, panel number)
plt.bar(df['University Name'], df['Fees'], color='lightblue')
plt.title('University Fees')
plt.xlabel('University Name')
plt.ylabel('Fees ($)')
plt.xticks(rotation=45)

# Bar plot for Scholarships
plt.subplot(1, 2, 2)
plt.bar(df['University Name'], df['Scholarship'], color='lightgreen')
plt.title('University Scholarships')
plt.xlabel('University Name')
plt.ylabel('Scholarship ($)')
plt.xticks(rotation=45)

# Show the plots
plt.tight_layout()
plt.show()
```

```python


# Rename columns
df.rename(columns={
    'University Name': 'USA University list',
    'Date of Birth': 'DOB',
    # Add more as needed
}, inplace=True)

# Alternatively, if you want to replace all column names
# df.columns = ['NewColumnName1', 'NewColumnName2', ...]  # Ensure the number matches

# Display the updated columns
print("Updated Columns:")
print(df.columns)

# Display the DataFrame
print(df)
```

```python


# Set the size of the plots
plt.figure(figsize=(10, 6))

# Pie chart for University Name
plt.subplot(1, 2, 1)  # (rows, columns, panel number)
df['University Name'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Universities')
plt.ylabel('')  # Hide y-label

# Pie chart for Country
plt.subplot(1, 2, 2)
df['Country'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Countries')
plt.ylabel('')  # Hide y-label

# Show the pie charts
plt.tight_layout()
plt.show()
```

```python
# Bar chart for average fees
df.groupby('University Name')['Fees'].mean().plot.bar()
plt.title('Average Fees per University')
plt.show()
```

```python
# Custom color palette
colors = ['gold', 'lightcoral', 'lightskyblue']
df['University Name'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('University Distribution')
plt.show()
```

```python
#Annotations for Insights
plt.pie(df['University Name'].value_counts(), autopct='%1.1f%%', startangle=90)
plt.text(0, 0, 'Key Insights', ha='center', va='center', fontsize=12, color='black')
plt.title('Distribution of Universities with Insights')
plt.show()
```

```python
# Create a comparative pie chart for different countries
plt.subplot(1, 2, 1)
df['Country'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Countries')

plt.subplot(1, 2, 2)
df[df['Scholarship']]['Country'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Scholarship Recipients by Country')
plt.show()

```
